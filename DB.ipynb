```python
import urllib.request
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
page = urllib.request.urlopen('https://docs.python.org/3/library/random.html')
```

```python
page.status
```

```python
soup = bs(page)
```

```python
print(soup.prettify())
```

```python
name = soup.body.findAll('dt')
name
```

```python
function_names = re.findall('id="random.\w+',str(name))
```

```python
new = []
for id in soup.find_all('dt'):
    if id.find('id') != None:
        rand = id.find('id')
        new.append(rand.attrs['random'])
        print(new)
```
